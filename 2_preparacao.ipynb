```python
#Importando bibliotecas


import mysql.connector
from mysql.connector import Error
import pandas as pd
```

***Conexão com o Banco de Dados Mysql***

```python
#!mysql -e "ALTER USER 'root'@'localhost' IDENTIFIED WITH 'mysql_native_password' BY 'root';FLUSH PRIVILEGES;"

# Create a connection to the MySQL server
conn = mysql.connector.connect(user='root', password='root', host='localhost')
# Create a cursor to interact with the MySQL server
cursor = conn.cursor()
```

```python
#Disponibilizando o DB Base de dados
cursor.execute("USE Base_dados")
```

```python
#Uma função para conectar ao nosso MySQL Server e ao banco de dados

def conecta_db(host, user,password,db):
    connection = None
    try:
        cursor.execute("USE Base_dados")
        conn = mysql.connector.connect(user='root', password='root', host='localhost', database='Base_dados')
        cursor = conn.cursor()
        print("Conexão bem sucedida")
    except Error as err:
        print(f"Erro : '{err}'")

    return connection

#Função para criar consultas

def executa_query(conn, query):
    cursor = conn.cursor()
    try:
        cursor.execute(query)
        print("Consulta feita com sucesso")
    except Error as err:
        print(f"Erro: '{err}'")

#Função para ler dados
def ler_query(conn, query):
    cursor = conn.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")
```

***2.1 - Importação e tratamento da tabela clientes***

```python
tab_clientes = """
SELECT *
FROM tab_clientes;
"""

connection = conecta_db("localhost","root","root","Base_dados")
results = ler_query(connection, tab_clientes)

#Tranformando a query em dataframe
df_clientes = pd.DataFrame(results)
df_clientes.head(10)
```
